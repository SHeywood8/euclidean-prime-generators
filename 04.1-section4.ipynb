## Section 4.3 - Sequences

#### The first Euclid-Mullin sequence in 4k+3 generated by f(x)=4x-1

```python
def f(x):
    return 4*x - 1

EM3mod4 = []
for iter in range(7):
    a = math.prod(EM3mod4)
    n = f(a)
    facs = funcs.factor2(n)
    print(f'n={n} with factors {facs} which are {[x % 4 for x in facs]} modulo 4.')
    EM3mod4.append(min([x for x in facs if x % 4 == 3]))
print(EM3mod4)
```

#### The first Euclid-Mullin sequence in 4k+3 generated by g(x)=4x+3

```python
def g(x):
    return 4*x + 3

EM3mod42 = []
for iter in range(9):
    a = math.prod(EM3mod42)
    n = g(a)
    facs = funcs.factor2(n)
    print(f'n={n} with factors {facs} which are {[x % 4 for x in facs]} modulo 4.')
    EM3mod42.append(min([x for x in facs if x % 4 == 3]))
print(EM3mod42)
```

```python
def g5(x):
    return 24*(432*(x**4) + 144*(x**3) + 18*(x**2) + x)+5

EM5mod24 = []
for iter in range(3):
    a = math.prod(EM5mod24)
    n = g5(2*(a**4))
    facs = funcs.factor2(n)
    print(f'n={n} with factors {facs} which are {[x % 24 for x in facs]} modulo 24.')
    EM5mod24.append(min([x for x in facs if x % 24 == 5]))
print(EM5mod24)
```

```python
n = g5(2*(math.prod(EM5mod24)**4))
print(len(str(n)))
```

```python
def g23(x):
    return 24*(432*(x**4) + 432*(x**3) + 150*(x**2) + 21*x) + 23

EM23mod24 = []
for iter in range(1):
    a = math.prod(EM23mod24)
    n = g23(2*(a**23))
    facs = funcs.factor2(n)
    print(f'n={n} with factors {facs} which are {[x % 24 for x in facs]} modulo 24.')
    EM23mod24.append(min([x for x in facs if x % 24 == 23]))
print(EM23mod24)
```

```python
n = g23(2*(math.prod(EM23mod24)**23))
print(len(str(n)))
```

#### 4 (mod 15)

```python
def f415(y):
    return y**4 - y**3 + 2*(y**2) + y + 1

EM4mod15 = []
for iter in range(5):
    a = math.prod(EM4mod15)
    n = f415((15*a) + 1)
    facs = funcs.factor2(n)
    print(f'n={n} with factors {facs} which are {[x % 15 for x in facs]} modulo 15.')
    EM4mod15.append(min([x for x in facs if x % 15 == 4]))
print(EM4mod15)
```

#### 6 (mod 7)

```python
def f67(x):
    return x**3 - 157*(x**2) + 8102*x - 137257

EM6mod7 = []
for iter in range(4):
    a = math.prod(EM6mod7)
    n = f67(7*a)
    facs = funcs.factor2(n)
    print(f'n={n} with factors {facs} which are {[x % 7 for x in facs]} modulo 7.')
    EM6mod7.append(min([x for x in facs if x % 7 == 6]))
print(EM6mod7)
```
